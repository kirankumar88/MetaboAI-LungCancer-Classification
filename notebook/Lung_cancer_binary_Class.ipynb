# Hyperparameter Tuning — Logistic Regression

```python
# Hyperparameter Tuning — Logistic Regression

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000, class_weight='balanced'))
])

param_grid = {
    'model__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'model__penalty': ['l2']
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=cv,
    scoring='roc_auc',
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV ROC-AUC:", grid.best_score_)
```

```python
# ============================================
# ML Biomarkers → KEGG Pathway Enrichment
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bioservices import KEGG
from scipy.stats import hypergeom

# ============================================
# Step 1 — Get Top Metabolites from Logistic Model
# ============================================

coefficients = ht_log_model.named_steps['model'].coef_[0]

importance_df = pd.DataFrame({
    "Metabolite": X.columns,
    "Coefficient": coefficients
})

importance_df["Abs_Coefficient"] = importance_df["Coefficient"].abs()
importance_df = importance_df.sort_values("Abs_Coefficient", ascending=False)

top_metabolites = importance_df.head(30)["Metabolite"].tolist()

print("Top metabolites:")
print(top_metabolites)

# ============================================
# Step 2 — Map Metabolites to KEGG IDs
# ============================================

k = KEGG()
met_kegg = []

for met in top_metabolites:
    try:
        res = k.find("compound", met)
        if res:
            kid = res.split("\t")[0].split(":")[1]
            met_kegg.append([met, kid])
    except:
        continue

met_kegg_df = pd.DataFrame(met_kegg, columns=["Metabolite","KEGG_ID"])
print("\nMetabolite → KEGG IDs")
print(met_kegg_df)

# ============================================
# Step 3 — Map KEGG IDs to Pathways
# ============================================

pathway_map = {}

for kid in met_kegg_df["KEGG_ID"]:
    try:
        links = k.link("pathway", "cpd:" + kid)
        if links:
            for line in links.strip().split("\n"):
                pathway_id = line.split("\t")[1].split(":")[1]
                pathway_map.setdefault(pathway_id, []).append(kid)
    except:
        continue

print("\nNumber of pathways found:", len(pathway_map))

# ============================================
# Step 4 — Convert Pathway IDs to Names
# ============================================

pathway_names = {}

for pid in pathway_map.keys():
    try:
        info = k.get(pid)
        name_line = [line for line in info.split("\n") if "NAME" in line][0]
        pathway_name = name_line.replace("NAME", "").strip()
        pathway_names[pid] = pathway_name
    except:
        pathway_names[pid] = pid

# ============================================
# Step 5 — Pathway Enrichment (Hypergeometric Test)
# ============================================

# Background = all metabolites in dataset
M = len(X.columns)

# Selected metabolites (mapped KEGG)
N = len(met_kegg_df["KEGG_ID"])

results = []

for pid, compounds in pathway_map.items():
    
    pathway_name = pathway_names[pid]
    pathway_compounds = set(compounds)
    
    k_hits = len(pathway_compounds & set(met_kegg_df["KEGG_ID"]))
    n = len(pathway_compounds)
    
    if k_hits > 0:
        pval = hypergeom.sf(k_hits-1, M, n, N)
        results.append([pathway_name, k_hits, pval])

enrichment_df = pd.DataFrame(results, columns=["Pathway","Hits","p-value"])
enrichment_df = enrichment_df.sort_values("p-value")

print("\nKEGG Pathway Enrichment Results:")
print(enrichment_df)

# Save results
enrichment_df.to_csv("KEGG_Pathway_Enrichment.csv", index=False)

# ============================================
# Step 6 — Bar Plot
# ============================================

top10 = enrichment_df.head(10)

plt.figure(figsize=(8,5))
plt.barh(top10["Pathway"], top10["Hits"])
plt.xlabel("Metabolite Hits")
plt.title("Top KEGG Pathways")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# ============================================
# Improved Bubble Plot — KEGG Pathway Enrichment
# ============================================

top10 = enrichment_df.head(10).sort_values("p-value")

plt.figure(figsize=(9,6))

y_pos = range(len(top10))

plt.scatter(
    top10["Hits"],
    y_pos,
    s=top10["Hits"]*150,
    c=-np.log10(top10["p-value"]),
    cmap="plasma",
    edgecolors="black"
)

plt.yticks(y_pos, top10["Pathway"])
plt.xlabel("Metabolite Hits")
plt.ylabel("KEGG Pathway")
plt.title("KEGG Pathway Enrichment Bubble Plot")

cbar = plt.colorbar()
cbar.set_label("-log10(p-value)")

plt.tight_layout()
plt.show()

```

```python
# Logistic Regression Hyperparameter Tuning (GridSearchCV)

from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('feature_selection', SelectKBest(f_classif)),
    ('model', LogisticRegression(max_iter=1000, class_weight='balanced'))
])

param_grid = {
    'feature_selection__k': [20, 30, 50, 70, 100],
    'model__C': [0.01, 0.1, 1, 10, 100],
    'model__penalty': ['l2'],
    'model__solver': ['lbfgs', 'liblinear']
}

grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best ROC-AUC:", grid.best_score_)
```

```python
# Evaluate Tuned Model

best_model = grid.best_estimator_

y_pred = best_model.predict(X_test)
y_prob = best_model.predict_proba(X_test)[:,1]

from sklearn.metrics import classification_report, roc_auc_score

print(classification_report(y_test, y_pred))
print("Test ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Tuned Logistic Regression")

plt.tight_layout()
plt.savefig("Confusion_Matrix_Tuned_LogReg.jpeg", dpi=300)
plt.savefig("/results/confusion_matrix.jpeg", dpi=300)
plt.show()
```

```python
# ROC_AUC curve

from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(y_test, y_prob)
auc = roc_auc_score(y_test, y_prob)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Tuned Logistic Regression")
plt.legend()

plt.tight_layout()
plt.savefig("ROC_Curve_Tuned_LogReg.jpeg", dpi=300)
plt.savefig("/results/ROC_curve.jpeg", dpi=300)
plt.show()
```

```python
# Precision recall curve
from sklearn.metrics import precision_recall_curve

precision, recall, _ = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(6,5))
plt.plot(recall, precision)

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")

plt.tight_layout()
plt.savefig("PR_Curve_Tuned_LogReg.jpeg", dpi=300)
plt.show()
```
